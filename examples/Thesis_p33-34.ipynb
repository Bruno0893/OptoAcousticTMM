```python
import numpy as np
from SuperlatticeTMM.raman import *
from SuperlatticeTMM.structure import StructureOA
import matplotlib.pyplot as plt
from materials import GaAs,AlAs,Air
```

```python
wl_DBR=900
params = [("GaAs",wl_DBR/4/GaAs.n),("AlAs",wl_DBR/4/AlAs.n)]*12 + [("GaAs",wl_DBR/GaAs.n)] + [("AlAs",wl_DBR/4/AlAs.n),("GaAs",wl_DBR/4/GaAs.n)]*12
params = np.array(params, dtype=[('materials','U10'),('thickness',np.float64)])
params
```

```python
struct = StructureOA(params, alloys=(0,0))
```

```python
wl_range = np.arange(700,1100,1)
plt.figure(figsize=(8,4))
plt.plot(wl_range,struct.get_optic_refl(wl_range,"GaAs","GaAs"),c='brown')
plt.xlabel("Wavelength(nm)")
plt.ylabel("Opt.Reflectivity")
```

```python
struct.plot_field("optic","real",900,"GaAs","GaAs",h=9)
```

```python
wl_range = np.arange(700,1100,1)
plt.figure(figsize=(8,5))
plt.plot(wl_range,struct.get_optic_refl(wl_range,"Air","GaAs"),c='brown')
plt.xlabel("Wavelength(nm)")
plt.ylabel("Opt.Reflectivity")
```
