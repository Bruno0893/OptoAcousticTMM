```python
import numpy as np
import matplotlib.pyplot as plt
from materials import GaAs,AlAs,Air,AlGaAsx
from SuperlatticeTMM.structure import StructureOA
```

```python
wl_DBR=915
GaAs_mod = AlGaAsx(0.1)
AlAs_mod = AlGaAsx(0.95)
params = [["GaAs",wl_DBR/4/GaAs_mod.n],["AlAs",wl_DBR/4/AlAs_mod.n]]*18 +[["GaAs",wl_DBR/2/GaAs_mod.n]] + \
[["GaAs",wl_DBR/4/GaAs_mod.n],["AlAs",wl_DBR/4/AlAs_mod.n]][::-1]*14
params
```

```python
struct = StructureOA(params,alloys=(0.1,0.95))
```

```python
wl_range= np.arange(780,1020,1)
plt.figure(figsize=(8,6))
plt.plot(wl_range,struct.get_optic_refl(wl_range,"GaAs","Air"),'r')
plt.xlabel("Wavelength(nm)",fontsize=12)
plt.ylabel("Opt.Reflectivity",fontsize=12)
```

```python
freq_range= np.arange(15,21,0.01)
plt.figure(figsize=(8,6))
plt.plot(freq_range,struct.get_acoustic_refl(freq_range,"GaAs","Air"),'b')
plt.xlabel("Frequency(GHz)",fontsize=12)
plt.ylabel("Ac.Reflectivity",fontsize=12)
```
